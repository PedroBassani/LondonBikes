```python
# Making a dictionary for new colunms names
new_cols_dict ={
    'timestamp':'date_time',
    'cnt':'count', 
    't1':'temp_real_C',
    't2':'temp_feels_like_C',
    'hum':'humidity_percent',
    'wind_speed':'wind_speed_kph',
    'weather_code':'weather',
    'is_holiday':'is_holiday',
    'is_weekend':'is_weekend',
    'season':'season'
}

# Renaming the columns to the specified column names
df.rename(new_cols_dict, axis=1, inplace=True)
```

```python
df.head()
```

```python
#
season_dict = {
    '0.0':'Spring',
    '1.0':'Summer',
    '2.0':'Fall',
    '3.0':'Winter'
}
df.season = df.season.astype('str')
df.season = df.season.map(season_dict)
```

```python
#
weather_dict = {
    '1.0':'Clear'
    , '2.0':'Scattered clouds'
    , '3.0':'Broken clouds'
    , '4.0':'Cloudy'
    , '7.0':'Rain'
    , '10.0':'Rain with thunderstorm'
    , '26.0':'Snowfall'
}

# changing the weather column data type to string
df.weather = df.weather.astype('str')
df.weather = df.weather.map(weather_dict)
```

```python
boolean_dict = {
    '0.0': 'False'
    , '1.0': 'True'
}
# Convertendo a coluna holiday para string e mapeando os valores
df.is_holiday = df.is_holiday.astype('str')
df.is_holiday = df.is_holiday.map(boolean_dict)
df.is_weekend = df.is_weekend.astype('str')
df.is_weekend = df.is_weekend.map(boolean_dict)
```
